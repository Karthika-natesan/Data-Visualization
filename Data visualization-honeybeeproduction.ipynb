## <font color = 'red'> *Question 3:*

### <font color = 'green'>*Display the percentage distribution of the data in each year using the pie chart.*


```python
# Using pie plot for analysing the distibution of the data year wise

labels = ['1998', '1999','2000','2001', '2002','2003','2004', '2005','2006','2007','2008',
       '2009','2010', '2011', '2012']

plt.pie(df.year.value_counts(), autopct = '%.1f%%', radius = 1.2, labels = labels)
plt.title("The percentage distribution of the data in each year \n")
# displaying the chart
plt.show()
```

#### Inferences from pie chart :
* From 1998 to 2000 , the honey production supply percentage is constant (i.e, 7.0% ), which is the highest in the trend line.
* From 2001 to 2003 , the honey production supply percentage is constant (i.e, 6.9%)
* From 2004 to 2008, the honey production supply percentage is constant(i.e, 6.5%)
* From 2009 to 2012, the honey production supply percentage is constant (i.e, 6.4%)

Henceforth, we can understand,how the percentage of honey production decreased over the preceeding years.

## <font color = 'red'> *Question 5:*

### <font color = 'green'>*Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.*


```python
# Analyzing the relationship between the variables 'numcol' and 'prodval' through scatterplot
x = df['numcol']
y = df['prodvalue']
plt.scatter(x,y,c='green')
# Assigning titles and labels to the plot
plt.title("Analysing the relationship between the variables 'numcol' and 'prodvalue'")
plt.xlabel('numcol')
plt.ylabel('prodvalue')
# generating a grid view 
plt.grid(True)
# Displaying the plot
plt.show()
```

#### Inferences observed from the scatterplot:
* The relationship between Number of honey producing colonies and Value of production from a scatter plot shows an upward trend,with points becoming more scattered as the values increase.
* Majority of the entries are within the range 0 to 100000 of numcol with prodvalue range 0 to 1.
* This explaines us that as the number of colonies increases, the production value tends to increase as well.

## <font color = 'red'> *Question 7:*

### <font color = 'green'>*Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')*

```python
# Analysing the relationship between the multiple pairs of variables throughout different years using pairplot 

# setting the figure size
sns.set(rc={"figure.figsize":(10,10)})

# Generating a pair plot
sns.pairplot(data= df,vars=['numcol', 'yieldpercol', 'totalprod', 'prodvalue'],hue='year', palette = "hls")

# display the plot
plt.show()
```

#### Inferences observed based on the plot:
* The relationship between 'numcol' and 'totalprod' is strongly positive and in upward trend, which indicates that as the number of colonies increases, the total production of honey also increases.
* The relationship between 'numcol' and 'prodvalue' is also strongly positive and in upward trend, which indicates that as the number of colonies increases, productionvalue of honey also increases.
* The relationship between 'yieldpersol' and 'totalprod' is either mildly positive or no relation because the points are scattered all around the plot.
* The diagonal of the pairplot shows a univariate distribution of the attributes.

## <font color = 'red'> *Question 8:*

### <font color = 'green'>*Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol','total prod', 'stocks', 'price per lb', 'prodvalue')*


```python
# creating a correlation matrix for the given variables
corr_matrix = df[['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']].corr()

# ploting a heatmap to view the correlation between the given variables 
sns.heatmap(corr_matrix, annot = True)
# display the plot
plt.show()
```

#### Inferences observed based on the plot:
* The values range from -1 to 1, where -1 indicates a perfect negative correlation, 0 indicates no correlation and 1 indicates a perfect positive correlation.

There is a strong positive correlation between :
* 'numcol' and 'totalprod'(0.95), which states that a higher number of colonies would produce more honey.
* 'totalprod' and 'prodvalue' (0.91), which states that more honey produced would result in a higher production value.
* 'numcol' and 'prodvalue' (0.91), which states that a higher number of colonies would result in a higher production value.

There is a strong negative correlation between:
* 'priceperlb' is negatively correlated with 'numcol', 'yeildpersol', 'totalprod','stocks' and 'prodvalue', which makes sense since an increase in production would lead to a excessive amount of honey, decreasing its price.
